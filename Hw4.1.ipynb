```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.interpolate import interp1d

speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
time = np.linspace(0, 11, 12)
print('Time: ', time)

plt.figure(figsize = (9, 5))
plt.plot(time, speed, 'o')
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
plt.xticks(time)

f1 = interp1d(time, speed, kind = 'cubic')
more_time = np.linspace(0, 11, 10000)
plt.plot(more_time, f1(more_time))

int_cub = integrate.quad(lambda x: f1(x), 0, 11)
print('Integral from cubic function: ', int_cub)
```

```python
plt.figure(figsize = (9, 5))
plt.plot(time, speed, 'o')
plt.xlim(0, 11)
plt.ylim(0, 130)
plt.grid()
plt.xticks(time)

f2 = interp1d(time, speed, kind = 'quadratic')
more_time = np.linspace(0, 11, 10000)
plt.plot(more_time, f2(more_time))

int_quad = integrate.quad(lambda x: f2(x), 0, 11)
print('Integral from quadratic function: ', int_quad)
```
