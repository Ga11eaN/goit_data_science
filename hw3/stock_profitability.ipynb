```python
import matplotlib.pyplot as plt
from numpy import median
import pandas as pd
import seaborn as sns

moment = [1,2,3,4,5,6]
profit_a = [25,-10,10,5,35,13]
profit_b = [0,15,-5,5,20,25]
profit_c = [10,25,-15,-5,-5,15]

df = pd.DataFrame({'moment': moment, 'profit_a': profit_a, 'profit_b': profit_b, 'profit_c':profit_c})

fig, axs = plt.subplots(ncols = 6, figsize = (15, 5))

sns.regplot(x = 'profit_a', y = 'profit_b', data = df, ax = axs[0])
sns.regplot(x = 'profit_b', y = 'profit_a', data = df, ax = axs[1])
sns.regplot(x = 'profit_b', y = 'profit_c', data = df, ax = axs[2])
sns.regplot(x = 'profit_c', y = 'profit_b', data = df, ax = axs[3])
sns.regplot(x = 'profit_a', y = 'profit_c', data = df, ax = axs[4])
sns.regplot(x = 'profit_c', y = 'profit_a', data = df, ax = axs[5])

#наименьшая зависимость между а и с
```
